This code builds a simple neural network that can recognize handwritten 28x28 pixel digits.

Our first version will have 15 neurons in the hidden layer and 10 output neurons.

See image below:
http://neuralnetworksanddeeplearning.com/images/tikz12.png

How many weights in this network?
11760 weights in the first layer (784 * 15) and 150 weights in the second layer (15 * 10) = 11910 weights total
15 biases in the hidden layer and 10 biases in the output layer = 25 biases total.
11910 + 25 = 11935 parameters!

```python
import nn.network as network
import mnist_loader
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
training_data = list(training_data)
```

```python
import numpy as np
print(np.shape(training_data))
print(np.shape(training_data[0][0]))
print(np.shape(training_data[0][1]))
print(training_data[0][1])
```

```python

# avg digit darkness

# Step 1 - function that calc. darkness from image
def img_darkness(imagepixeldata):
  darkness_total = 0
  for x in imagepixeldata:
    darkness_total += x
  return darkness_total[0]

# Step 2 - call function on training data
darkness = img_darkness(training_data[0][0])
whichimage = training_data[0][1]
whichimage_pos = np.argmax(whichimage)
print(f"Example: Image {whichimage_pos} Darknes: {darkness}")
#there are ten total darkness values one for each integer (0-9)

def print_data(data, pos):
  darkness = img_darkness(data[pos][0])
  whichimage = training_data[pos][1]
  whichimage_pos = np.argmax(whichimage)
  print(f"Image {whichimage_pos} Darknes: {darkness}")
```

```python
# Step 3 - use those averages to classify all data

# Gets multiple darkness values for every image, 0-9
darkness_counts = [[] for _ in range(10)]

# totals the darkness per image classification, then divides it by the length of it's list for an average.
darkness_avg = [0 for _ in range(10)]


# for img in training_data[:50]: # limit on training data for quicker computation and debugging
for img in training_data:
  darkness = img_darkness(img[0])
  image_num = np.argmax(img[1])
  darkness_counts[image_num].append(darkness)
print(darkness_counts)

index = 0
for rows in darkness_counts:
  print(index, rows)
  if len(rows) != 0:
    for cols in rows:
      darkness_avg[index] += cols
    darkness_avg[index] = darkness_avg[index] / float(len(rows))
  index += 1

print()

for i in range(10):
  print(f"Avgerage darkness of {i}: {darkness_avg[i]}")
```
